```python
# type your code here
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
import seaborn as sns
```

```python
# Assuming df is your DataFrame after transforming numerical variables
# Dropping the 'Region' column for clustering
X = df.drop('Region', axis=1)
```

```python
silhouette_avg = []
sample_silhouette_values = []
```

```python
for n_clusters in range(2, 5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    
    # Compute the silhouette score
    silhouette_avg.append(silhouette_score(X, cluster_labels))
    
    # Compute silhouette samples
    sample_silhouette_values.append(silhouette_samples(X, cluster_labels))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 5), silhouette_avg, marker='o', linestyle='-', color='b')
plt.title('Average Silhouette Score for Different K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Average Silhouette Score')
plt.xticks(range(2, 5))
plt.grid()
plt.show()
```

```python
for i, n_clusters in enumerate(range(2, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(f'Silhouette Plot for K={n_clusters}')
    
    silhouette_vals = sample_silhouette_values[i]
    y_lower = 10  # Starting y position for the plot
    
    sorted_silhouette_vals = np.sort(silhouette_vals)
    size_cluster = np.bincount(cluster_labels)  # Get sizes of clusters
    for j in range(n_clusters):
        size = size_cluster[j]
        y_upper = y_lower + size
        
        plt.fill_betweenx(np.arange(y_lower, y_upper), 0, sorted_silhouette_vals[:size])
        plt.text(-0.05, y_lower + 0.5 * size, str(j))
        y_lower = y_upper  # Update y lower bound for next cluster
    
    plt.axvline(x=silhouette_avg[i], color='red', linestyle='--')
    plt.xlim([-0.1, 1])
    plt.xlabel('Silhouette Coefficient')
    plt.ylabel('Cluster Label')
    plt.title(f'Silhouette Plot for K={n_clusters}')
    plt.show()
```

```python
# type your code here
# Standardize the numerical variables
scaler = StandardScaler()
numerical_vars = ['Vegetables', 'Personal_care']
df_scaled = scaler.fit_transform(df[numerical_vars])

# Fit KMeans clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Visualize the clusters using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Vegetables', y='Personal_care', hue='Cluster', palette='viridis', style='Cluster', markers=['o', 'X'])
plt.title('K-Means Clustering (2 Clusters) based on Vegetables and Personal Care')
plt.xlabel('Vegetables (Standardized)')
plt.ylabel('Personal Care (Standardized)')
plt.legend(title='Cluster')
```

```python
# Interpretation
print("Cluster Summary Statistics:")
cluster_summary = df.groupby('Cluster')[numerical_vars].agg(['mean', 'std', 'min', 'max', 'count'])
print(cluster_summary)

# Additional Insights
for cluster in cluster_summary.index:
    print(f"\nCluster {cluster} Analysis:")
    print(f" - Mean Vegetables: {cluster_summary.loc[cluster, ('Vegetables', 'mean')]:.2f}")
    print(f" - Mean Personal Care: {cluster_summary.loc[cluster, ('Personal_care', 'mean')]:.2f}")
    print(f" - Number of observations: {cluster_summary.loc[cluster, ('Vegetables', 'count')]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset
#df = pd.read_csv(r"H:\My Drive\MACHINE LEARNING+GL\ML+USL+GL\DAY 1\USL - Take home (Day 1) - Questions-1\USL - Take home (Day 1) - Questions\wholesale_cust.csv")

# Step 1: Filter for Oneonta
oneonta_data = df[df['Region'] == 'Oneonta'].drop(columns=['Region'])

# Step 2: Standardize the data
scaler = StandardScaler()
oneonta_scaled = scaler.fit_transform(oneonta_data)

# Step 3: Calculate WCSS for K values from 1 to 10 (for better analysis)
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(oneonta_scaled)
    wcss.append(kmeans.inertia_)  # WCSS is stored in inertia_

# Step 4: Plot WCSS against K
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='b')
plt.title('Optimal Number of Clusters Using WCSS')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within Cluster Sum of Squares (WCSS)')
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```
