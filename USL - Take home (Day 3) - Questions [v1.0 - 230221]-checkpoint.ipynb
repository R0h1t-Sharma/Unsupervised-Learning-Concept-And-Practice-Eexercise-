```python
import numpy as np
import matplotlib.pyplot as plt

# Given eigenvalues
eigenvalues = np.array([3.31384109, 2.13929696, 1.20400993, 1.10031049, 
                        0.97664776, 0.87625729, 0.78373939, 0.72093906, 
                        0.46002434, 0.22201733, 0.17782633, 0.04457056])

# Calculate explained variance
explained_variance = eigenvalues / np.sum(eigenvalues)

# Create scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.xticks(range(1, len(eigenvalues) + 1))
plt.grid()
plt.show()

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance)

# Print cumulative explained variance for optimal components
for i, var in enumerate(cumulative_variance):
    print(f'Components: {i + 1}, Cumulative Explained Variance: {var:.2f}')
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame containing the relevant numerical data
# Standardizing the data
features = df.select_dtypes(include=[np.number])  # Select only numerical columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

# Step 1: Compute the covariance matrix
cov_matrix = np.cov(scaled_data, rowvar=False)

# Step 2: Compute the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 3: Extract the largest eigenvalue and corresponding eigenvector
largest_index = np.argmax(eigenvalues)
lambda_max = eigenvalues[largest_index]
x = eigenvectors[:, largest_index]

# Step 4: Check if Ax = λx
Ax = cov_matrix.dot(x)
lambda_x = lambda_max * x

# Check equality
is_equal = np.allclose(Ax, lambda_x)

# Print results
print(f'Largest Eigenvalue (λ): {lambda_max}')
print(f'Corresponding Eigenvector (x): {x}')
print(f'Ax: {Ax}')
print(f'λx: {lambda_x}')
print(f'Are Ax and λx equal? {is_equal}')
```
